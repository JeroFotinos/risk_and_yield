# Imports

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data", "Wheather")

print(DATA_PATH)
```

# PAR

## 1. 2020-12-01 to 2021-04-30

```python
df_par = pd.read_csv(DATA_PATH / "ClimaPAR.csv")
```

```python
df_par["scene_date"] = pd.to_datetime(
    df_par[["AÑO", "MES ", "DIA"]].rename(
        columns={"AÑO": "year", "MES ": "month", "DIA": "day"}
    )
)

df_par[["scene_date", "PAR"]]
```

## 2. 2021-11-30 to 2022-01-16	

```python
df_par = pd.read_excel(DATA_PATH / "PAR 301121 al 170122.xlsx")
df_par["scene_date"] = pd.to_datetime(
    df_par[["YEAR", "MES", "DIA"]].rename(
        columns={"YEAR": "year", "MES": "month", "DIA": "day"}
    )
)
```
